```python
import os
import urllib
import cv2
import numpy as np
from pixellib.semantic import semantic_segmentation
from matplotlib import pyplot as plt
```

```python
img_path = os.getenv('HOME')+'/aiffel/human_segmentation/images/my_image.png'  
img_orig = cv2.imread(img_path) 

print(img_orig.shape)

# cv2.cvtColor( ,   ):     
# cv2.COLOR_BGR2RGB:     (BGR  RGB  )
# plt.imshow():     ,  RGB(A)   2D  
# https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.imshow.html
# plt.show():    figure  ( figure ,  )
# https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.show.html
plt.imshow(cv2.cvtColor(img_orig, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
model_dir = os.getenv('HOME')+'/aiffel/human_segmentation/models' 
# 2
model_file = os.path.join(model_dir, 'deeplabv3_xception_tf_dim_ordering_tf_kernels.h5') 
```

```python
model = semantic_segmentation() #PixelLib      semantic segmentation    
model.load_pascalvoc_model(model_file) # pascal voc    (model_file)   
```

```python
segvalues, output = model.segmentAsPascalvoc(img_path)
```

```python
LABEL_NAMES = [
    'background', 'aeroplane', 'bicycle', 'bird', 'boat', 'bottle', 'bus',
    'car', 'cat', 'chair', 'cow', 'diningtable', 'dog', 'horse', 'motorbike',
    'person', 'pottedplant', 'sheep', 'sofa', 'train', 'tv'
]
len(LABEL_NAMES)
```

```python
#segmentAsPascalvoc()           output matplotlib  
plt.imshow(output)
plt.show()
```

```python
segvalues # segmentAsPascalvoc()           
```

```python
#segvalues  class_ids    pacalvoc   
for class_id in segvalues['class_ids']:
    print(LABEL_NAMES[class_id])
```

```python
#     
# PixelLib   
#      !

#  
colormap = np.zeros((256, 3), dtype = int)
ind = np.arange(256, dtype=int)

for shift in reversed(range(8)):
    for channel in range(3):
        colormap[:, channel] |= ((ind >> channel) & 1) << shift
    ind >>= 3

colormap[:20] # 20  
```

```python
colormap[15] # 15    (pacalvoc LABEL_NAMES 15 )
```

```python
seg_color = (128,128,192) #   - colormap  RGB  output  BGR     
```

```python
# output    seg_color  1(True),  0(False) 
# seg_color  person       gray 
# cmap      
seg_map = np.all(output==seg_color, axis=-1) 
print(seg_map.shape) 
plt.imshow(seg_map, cmap='gray')
plt.show()
```

```python
#  img_show         color_mask      
img_show = img_orig.copy()

# True False   255 0 
img_mask = seg_map.astype(np.uint8) * 255

# 255 0   
color_mask = cv2.applyColorMap(img_mask, cv2.COLORMAP_JET)

#     
# 0.6 0.4    .
img_show = cv2.addWeighted(img_show, 0.6, color_mask, 0.4, 0.0)

plt.imshow(cv2.cvtColor(img_show, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
# (13,13) blurring kernel size 
#  
img_orig_blur = cv2.blur(img_orig, (13,13))

# plt.imshow():     .
# cv2.cvtColor( ,   ):     
# cv2.COLOR_BGR2RGB:  BGR   
#     (BGR  RGB  )   
plt.imshow(cv2.cvtColor(img_orig_blur, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
# (13,13) blurring kernel size 
#  
img_orig_blur = cv2.blur(img_orig, (19,19))

# plt.imshow():     .
# cv2.cvtColor( ,   ):     
# cv2.COLOR_BGR2RGB:  BGR   
#     (BGR  RGB  )   
plt.imshow(cv2.cvtColor(img_orig_blur, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
#  blur   
```

```python
# cv2.cvtColor( ,   ):     
# cv2.COLOR_BGR2RGB:  BGR   
#     (BGR  RGB  ) 
img_mask_color = cv2.cvtColor(img_mask, cv2.COLOR_GRAY2BGR)

# cv2.bitwise_not():  .  0  255 
#     255  0.
img_bg_mask = cv2.bitwise_not(img_mask_color)

# cv2.bitwise_and()       .
# 0   bitwise_and   0   
#  0      0 .     !
img_bg_blur = cv2.bitwise_and(img_orig_blur, img_bg_mask)
plt.imshow(cv2.cvtColor(img_bg_blur, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
# np.where(, , )
#   255       
#      .
img_concat = np.where(img_mask_color==255, img_orig, img_bg_blur)
# plt.imshow():     .
# cv2.cvtColor( ,   ):     
# cv2.COLOR_BGR2RGB:  BGR    
#     (BGR  RGB  )
plt.imshow(cv2.cvtColor(img_concat, cv2.COLOR_BGR2RGB))
plt.savefig("man_blured.jpg", dpi = 300)
plt.show()
```

```python
def image_process(img_path2):
    img_orig2 = cv2.imread(img_path2) 

    print(img_orig2.shape)

    plt.imshow(cv2.cvtColor(img_orig2, cv2.COLOR_BGR2RGB))
    plt.show()
    
    segvalues2, output2 = model.segmentAsPascalvoc(img_path2)
    
    plt.imshow(output2)
    plt.savefig("my_1.jpg", dpi = 600)
    plt.show()
    
    print(segvalues2)
    
    for class_id in segvalues2['class_ids']:
        print(LABEL_NAMES[class_id])
        
    seg_color = (128,128,192)
    
    seg_map2 = np.all(output2==seg_color, axis=-1) 
    print(seg_map2.shape) 
    plt.imshow(seg_map2, cmap='gray')
    plt.savefig("my_2.jpg", dpi = 600)
    plt.show()
    
    img_show2 = img_orig2.copy()

    img_mask2 = seg_map2.astype(np.uint8) * 255

    color_mask2 = cv2.applyColorMap(img_mask2, cv2.COLORMAP_JET)

    img_show2 = cv2.addWeighted(img_show2, 0.6, color_mask2, 0.4, 0.0)

    plt.imshow(cv2.cvtColor(img_show2, cv2.COLOR_BGR2RGB))
    plt.savefig("my_3.jpg", dpi = 600)
    plt.show()
    
    img_orig_blur2 = cv2.blur(img_orig2, (17,17))

    plt.imshow(cv2.cvtColor(img_orig_blur2, cv2.COLOR_BGR2RGB))
    plt.show()
    
    img_mask_color2 = cv2.cvtColor(img_mask2, cv2.COLOR_GRAY2BGR)

    img_bg_mask2 = cv2.bitwise_not(img_mask_color2)

    img_bg_blur2 = cv2.bitwise_and(img_orig_blur2, img_bg_mask2)
    plt.imshow(cv2.cvtColor(img_bg_blur2, cv2.COLOR_BGR2RGB))
    plt.savefig("my_4.jpg", dpi = 600)
    plt.show()
    
    img_concat2 = np.where(img_mask_color2==255, img_orig2, img_bg_blur2)

    plt.imshow(cv2.cvtColor(img_concat2, cv2.COLOR_BGR2RGB))
    plt.savefig("my_5.jpg", dpi = 600)
    plt.show()
```

```python
image_process(os.getenv('HOME')+'/aiffel/human_segmentation/images/20231116_1.jpg')
```
